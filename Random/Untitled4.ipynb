```python
topdenied = df[df.CASE_STATUS == 'DENIED']['WORKSITE'].value_counts()[:10]
```

```python
fig = plt.figure(figsize=(15,10),facecolor='skyblue')
color = ['slateblue','darkviolet','violet','indigo','magenta','purple','darkorchid','darkviolet']

ax1 = fig.add_axes([0,0,0.45,0.45])
ax2 = fig.add_axes([0.5,0,0.45,0.45])

ax1.bar(x=topwork.index,height=topwork[:],color=color)
ax1.set_title('top 10 worksites')
ax1.set_xticklabels(labels=topwork.index,rotation=90)
ax1.set_xlabel('worksites name')
ax1.set_ylabel('Top 10 companies')

ax2.bar(x=topdenied.index,height=topdenied[:])
ax2.set_title('top 10 denied worksites')
ax2.set_xticklabels(labels=topdenied.index,rotation=90)
ax2.set_xlabel('worksites name')
ax2.set_ylabel('Top 10 companies')
```

```python
job = ['DATA SCIENTIST','DATA ANALYST','CEO','DATA ENGINEER']
total =[]
for var in job:
    d = df[df.JOB_TITLE == var]['JOB_TITLE'].count()
    total.append(d)
```

```python
plt.figure(figsize=(10,5),facecolor='red')
plt.bar(x=job,height=total[:],color=['m','r','g','y'])
plt.title('Data scientist overall posts')
plt.xlabel('Names of different positions')
plt.ylabel('Number of values of total jobs')
plt.show()
```
