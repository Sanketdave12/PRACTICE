```python
angles = [np.deg2rad(var) for var in range(361)]
```

```python
angles[:20]
```

```python
plt.figure(figsize=(12,6),facecolor='seagreen')
plt.plot(angles,np.sin(angles),color='yellow',label='$Sin \Theta$')
plt.plot(angles,np.cos(angles),color='green',label='$Cos \Theta$')
plt.title('plot of radians with sin and cos')
plt.legend()
plt.xticks(range(0,7))
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.distplot(a['total_bill'], bins=30,
            hist_kws = {'alpha':1, 'histtype':'step' },
            kde_kws = {'alpha':0.3 , 'color':'red', 'lw':3},
            ax = ax ) # dist == histogram
plt.show()
```
