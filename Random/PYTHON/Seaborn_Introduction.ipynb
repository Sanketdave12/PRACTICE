**Distribution of tota_bill in tip dataset**

```python
# matplotlib
print(tip['total_bill'].describe())
plt.figure(figsize=(10,6))
plt.hist(tip['total_bill'], bins=30, ec='cyan')
plt.xlabel("Data Points", fontsize=20, color='red')
plt.ylabel("Frequency", fontsize=20, color='red')
plt.xticks(range(0,51,2))
plt.yticks(range(0,30,2))
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.distplot(tip['total_bill'], bins=30, kde=True,
            hist_kws = {'alpha':1, 'histtype':'step' },
            kde_kws = {'alpha':1 , 'color':'red', 'lw':3},
            ax = ax ) # dist == histogram
plt.show()
```

```python
sns.distplot(tip['total_bill'], bins=30, kde=False,
            hist_kws = {'alpha':1, 'ec':'cyan'}) # dist == histogram
plt.show()
```

**scatter plot** it shows releationship between two variables

```python
c = tip['sex'].map({'Female':'red', 'Male':'blue'})
plt.figure(figsize=(10, 4))
plt.scatter('total_bill', 'tip', data=tip, color=c)
plt.grid(color='#123456')
plt.xlabel('Total Bill', color='green', fontsize=20)
plt.ylabel('Tip', color='green', fontsize=20)
plt.xticks(range(0,55,5))
plt.yticks(range(12))
plt.show()
```
