```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import random
```

```python
norm = np.random.normal(1,80,1000)
```

```python
mean = np.mean(norm)
```

```python
median =np.median(norm)
```

```python
std = np.std(norm)
```

```python
std
```

```python
plt.figure(figsize=(10,5),facecolor='indianred')
plt.hist(norm,bins=5,color='black',edgecolor='orange')
plt.plot([mean,mean],[0,81],color='g',lw=4)
plt.plot([median,median],[0,81],color='y',lw=2)
plt.plot([std,std],[0,81],color='m')
plt.plot([mean-std,mean-std],[0,81],color='m')
plt.plot([2*(mean-std),2*(mean-std)],[0,81],'b')
plt.plot([2*(mean+std),2*(mean+std)],[0,81],'b')

plt.title("hehe")
plt.show()
```

```python
plt.figure(figsize=(10,5),facecolor='indianred')
plt.hist(norm,bins=5,color='black',edgecolor='orange')
plt.plot([mean,mean],[0,888],color='g',lw=4)
plt.plot([median,median],[0,888],color='y',lw=2)
plt.plot([std,std],[0,888],color='m')
plt.title("HEHE 2")
plt.show()
```
